Matplotlib example workflow

```python
# 0. import matplotlib and get it ready for plotting in jupyter
%matplotlib inline
import matplotlib.pyplot as plt
# 1. Prepare data
x = [1,2,3,4]
y = [11,22,33,44]

# 2. Setup plot
fig,ax = plt.subplots(figsize = (10,10)) # (width , height)

# 3. Plot data
ax.plot(x,y)

# 4. Customize plot
ax.set(title = "Simple plot",
       xlabel = "x-axis",
       ylabel = "y-axis")

# 5. Save and show (you save the whole figure)
fig.savefig("sample-plot.png")
```

# Two options for subplots

```python
# subplot option 1
fig, ((ax1,ax2),(ax3,ax4)) = plt.subplots(nrows = 2,
                                          ncols =2,
                                          figsize = (10,5))
# plot to each different axis
ax1.plot(x,x/2);
ax2.scatter(np.random.random(10),np.random.random(10));
ax3.bar(nut_butter_prices.keys(),nut_butter_prices.values());
ax4.hist(np.random.randn(1000));
```

```python
# subplot option 2
fig, ax = plt.subplots(nrows = 2,ncols =2,figsize = (10,5))
#Plot each different axes
ax[0][0].plot(x,x/2);#ax[0,0].plot(x,x/2)
ax[0,1].scatter(np.random.random(10),np.random.random(10));
ax[1,0].bar(nut_butter_prices.keys(),nut_butter_prices.values());
ax[1,1].hist(np.random.randn(1000));
```

# Which one should you use?(pyolot vs matplotlib  oo method?)

* When plotting something quickly okey use the plyplot method
* When potiing something more advanced, use the oo method

```python
heart_disease.head()
```

```python
over_50 = heart_disease[heart_disease["age"] > 50]
over_50.head()
```

```python
over_50.plot(kind="scatter",x="age",y="chol",c="target",figsize=(10,10));
```

```python
# OO method
fix, ax = plt.subplots(figsize=(10,10))
over_50.plot(kind="scatter",x="age",y="chol",c="target",ax =ax); #ax= ax means plotting in this axes
```

```python
fix,ax = plt.subplots(figsize=(10,10))
over_50.plot(kind="scatter",x="age",y="chol",ax = ax,c="target")
ax.set_xlim([50,100])# changing limits of axis i.e:X - axis
```

```python
over_50.target.values
```

```python
# oo method from scratch
fix, ax = plt.subplots(figsize = (10,6))

#Plot the data
scatter = ax.scatter(x=over_50["age"],
                     y=over_50["chol"],
                     c=over_50["target"])
# Customize the plot
ax.set(title = "Heart Disease and Cholesterol Levels",
       xlabel= "age",
       ylabel = "cholesterol")

# Add a legend
ax.legend(*scatter.legend_elements(),title ="target");# grab value from scatter elements & unpack legend_elements which is query "c" paramter  

# Add a horizontal line
ax.axhline(over_50["chol"].mean(),linestyle ="--");
```

```python
over_50.head()
```

```python
# Subplot of chol, age, thalach

fig, (ax0, ax1) = plt.subplots(nrows=2,ncols=1,figsize=(10,10))

# Add data to ax0

scatter = ax0.scatter(x=over_50["age"],
                      y=over_50["chol"],
                      c=over_50["target"]);

ax0.set(title = "Age and Cholestrol", 
        xlabel="age", 
        ylabel = "chol")

# Add legend
ax0.legend(*scatter.legend_elements(),title="target")

# Add mean line
ax0.axhline(y=over_50["chol"].mean(),linestyle="--");



# Add data to ax1

scatter = ax1.scatter(x=over_50["age"],
                      y=over_50["thalach"],
                      c=over_50["target"]);

ax1.set(title = "Age and thalach",# thalach rate of heart 
        xlabel="age", 
        ylabel = "thalach")

# Add legend
ax1.legend(*scatter.legend_elements(),title="target")

# Add mean line
ax1.axhline(y=over_50["thalach"].mean(),linestyle="--");


# Add title to Figure

fig.suptitle("Heart Disease Analysis", fontsize = 16,fontweight = "bold");
```

```python
# Subplot of chol, age, thalach
plt.style.use("seaborn-whitegrid")

fig, (ax0, ax1) = plt.subplots(nrows=2,ncols=1,figsize=(10,10))

# Add data to ax0

scatter = ax0.scatter(x=over_50["age"],
                      y=over_50["chol"],
                      c=over_50["target"],
                      cmap="RdYlBu"); # this changes the colour scheme

ax0.set(title = "Age and Cholestrol", 
        xlabel="age", 
        ylabel = "chol")
# change x limits with range from 50 to 80
ax0.set_xlim([50,80])
# ax0.set_ylim([60,200])
# Add legend
ax0.legend(*scatter.legend_elements(),title="target")

# Add mean line
ax0.axhline(y=over_50["chol"].mean(),linestyle="--");

# Add data to ax1

scatter = ax1.scatter(x=over_50["age"],
                      y=over_50["thalach"],
                      c=over_50["target"],
                      cmap = "winter");
ax1.set(title = "Age and thalach",# thalach rate of heart 
        xlabel="age", 
        ylabel = "thalach")
# change x limits with range from 50 to 80
ax1.set_ylim([50,200])
ax1.set_xlim([50,80])
# Add legend
ax1.legend(*scatter.legend_elements(),title="target")

# Add mean line
ax1.axhline(y=over_50["thalach"].mean(),linestyle="--");


# Add title to Figure

fig.suptitle("Heart Disease Analysis", fontsize = 16,fontweight = "bold");
```
