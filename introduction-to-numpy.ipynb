## Dot product (nut butter sales)

```python
np.random.seed(0)
# Number of jar sold
sales_amounts = np.random.randint(20, size = (5,3))
sales_amounts
```

```python
#Create weekly_sales DataFrame
weekly_sales = pd.DataFrame(sales_amounts,index= ["Mon","Tue","wed","Thur","Fri"],
                            columns=["Almond butter","Peanut butter","cashew butter"])
weekly_sales
```

```python
# Create prices array 
prices = np.array([10,8,12])
prices
```

```python
# Crete butter_prices DataFrame
butter_price = pd.DataFrame(prices.reshape(1,3),index=["Price"],
                            columns=["Almond butter","Peanut butter","cashew butter"])
butter_price
```

```python
prices
```

```python
sales_amounts.T.shape , prices.shape# this means (1,3)
```

```python
# shape aren't aligned, let's transposed
total_sales = prices.dot(sales_amounts.T)
total_sales
```

```python
butter_price.shape , weekly_sales.shape # adding two dataframe in to one
```

```python
# Create dailysales
daily_sales = butter_price.dot(weekly_sales.T) 
daily_sales
```

```python
weekly_sales
```

```python
# Doesnot work, Trying to fixing row and column, not right shape in column
# now it is fixed by exchanging row and column 
Transpose_value= daily_sales.T
Transpose_value
```

```python
weekly_sales["Price"] = Transpose_value
weekly_sales
```
